```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv', encoding='utf-8')
```

```python
train
```

```python
print(len(train))
train.isnull().sum()
```

```python
del train['인승']
del train['자동수동']
del train['no']
del train['연월']
```

```python
train.isnull().sum()
```

```python
deli = []
dist = []

for i, row in train.iterrows():
    if '만' in row['주행거리']:
        dist.append(int(row['주행거리'][:-3]) * 10000)
    elif '천' in row['주행거리']:
        dist.append(int(row['주행거리'][:-3]) * 1000)
    else:
        if row['주행거리'] == '등록':
            deli.append(i)
            continue
        dist.append(int(row['주행거리'][:-2]))
        
train.drop(deli, inplace=True)
del train['주행거리']
train.index = range(len(train))
print(len(train))

train['주행거리'] = dist
```
